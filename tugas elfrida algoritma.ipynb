```
while True:
    print("\n=== MENU ===")
    print("1. A pangkat B")
    print("2. Hitung deret pecahan")
    print("0. Keluar")

    pilihan = int(input("Masukkan pilihan anda: "))

    if pilihan == 1:
        bilangan = int(input("Masukkan bilangan bulat: "))
        pangkat = int(input("Masukkan pangkat yang diinginkan: "))

        hasil = 1
        for i in range(1, pangkat + 1):
            hasil = hasil * bilangan
            print(f"hasil {bilangan} pangkat {i} adalah {hasil}")

    elif pilihan == 2:
        n = int(input("Masukkan jumlah N: "))

        a, b = 1, 1  # Fibonacci awal
        hasil = 1    # suku pertama

        print("\nDeret:")
        print("1", end=" ")

        for i in range(1, n):
            c = a + b
            a, b = b, c

            pecahan = a / b

            if i % 2 == 1:
                print(f"- {a}/{b}", end=" ")
                hasil -= pecahan
            else:
                print(f"+ {a}/{b}", end=" ")
                hasil += pecahan

        print("\nHasil =", hasil)


    elif pilihan == 0:
        print("Program selesai")
        break

    else:
        print("Pilihan tidak valid")
```
